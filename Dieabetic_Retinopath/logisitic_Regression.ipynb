```python
# -------------------------------
# 1. Import Libraries
# -------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# 2. Load Dataset
# -------------------------------
data = pd.read_csv("diabetes.csv")
print(data.head())
print(data.info())
print(data.describe())

# -------------------------------
# 3. Check for Missing Values
# -------------------------------
print("Missing values per column:")
print(data.isnull().sum())

# -------------------------------
# 4. Correlation Heatmap
# -------------------------------
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap", fontsize=16, fontweight="bold")
plt.show()

# -------------------------------
# 5. Distribution of Target (Outcome)
# -------------------------------
sns.countplot(x="Outcome", data=data, palette="pastel")
plt.title("Distribution of Diabetes Outcomes", fontsize=16, fontweight="bold")
plt.xlabel("Outcome (0 = Non-Diabetic, 1 = Diabetic)")
plt.ylabel("Count")
plt.show()

# -------------------------------
# 6. Histograms of Features
# -------------------------------
data.hist(bins=30, figsize=(12,10), color="skyblue", edgecolor="black")
plt.suptitle("Feature Distributions", fontsize=18, fontweight="bold")
plt.show()

# -------------------------------
# 7. Histograms Split by Outcome
# -------------------------------
features = ["Pregnancies", "Glucose", "BloodPressure", "BMI", "Age"]
for col in features:
    plt.figure(figsize=(6,4))
    sns.histplot(data=data, x=col, hue="Outcome", kde=True, palette="Set2", bins=30, alpha=0.6)
    plt.title(f"Distribution of {col} by Outcome", fontsize=14, fontweight="bold")
    plt.show()

# -------------------------------
# 8. Boxplots (Example: BMI vs Outcome)
# -------------------------------
plt.figure(figsize=(6,4))
sns.boxplot(x="Outcome", y="BMI", data=data, palette="Set2")
plt.title("Boxplot of BMI by Outcome", fontsize=14, fontweight="bold")
plt.xlabel("Outcome (0 = Non-Diabetic, 1 = Diabetic)")
plt.ylabel("BMI")
plt.show()

# -------------------------------
# 9. Pairplot (Optional: Multivariate Visualization)
# -------------------------------
sns.pairplot(data[["Glucose", "BMI", "Age", "Pregnancies", "Outcome"]], hue="Outcome", palette="husl")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
```

```python
df.isnull().sum()
```

```python
X = df.drop("Outcome", axis=1)  # Features
y = df["Outcome"]               # Target
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:, 1]  # Probabilities for ROC curve
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))

```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label="AUC = "+str(round(roc_auc_score(y_test, y_pred_prob), 2)))
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_[0]
}).sort_values(by="Coefficient", ascending=False)

print(feature_importance)
```

```python
important_features = ['Glucose', 'BMI', 'Pregnancies', 'DiabetesPedigreeFunction', 'Age']
X_new = df[important_features]
y = df['Outcome']
```

```python
scaler = StandardScaler()
X_new_scaled = scaler.fit_transform(X_new)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_new_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

model_new = LogisticRegression(random_state=42)
model_new.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Predictions
y_pred = model_new.predict(X_test)
y_pred_prob = model_new.predict_proba(X_test)[:,1]

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Classification Report
print(classification_report(y_test, y_pred))

# ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label="AUC = "+str(round(roc_auc_score(y_test, y_pred_prob),2)))
plt.plot([0,1],[0,1],'--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
