```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris(as_frame=True)
dat = iris.frame
dat['Species'] = iris.target_names[iris.target]  # Add species column

# 1. Minimum, Maximum, and Range
min_val = dat['sepal length (cm)'].min()
max_val = dat['sepal length (cm)'].max()
range_val = max_val - min_val

# 2. Mean and Median
mean_val = dat['sepal length (cm)'].mean()
median_val = dat['sepal length (cm)'].median()

# 3. Quartiles and IQR
Q1 = dat['sepal length (cm)'].quantile(0.25)
Q3 = dat['sepal length (cm)'].quantile(0.75)
iqr_val = Q3 - Q1

# 4. Variance and Standard Deviation
var_val = dat['sepal length (cm)'].var()
sd_val = dat['sepal length (cm)'].std()

# 5. Summary Statistics
summary_stats = dat.describe()

# 6. Mode
mode_val = dat['sepal length (cm)'].mode()[0]
mode_freq = dat['sepal length (cm)'].value_counts().iloc[0]

# 7. Correlation
cor_val = dat['sepal length (cm)'].corr(dat['sepal width (cm)'])

# 8. Contingency Table
median_val = dat['sepal length (cm)'].median()
dat['size'] = np.where(dat['sepal length (cm)'] < median_val, "small", "big")
cont_table = pd.crosstab(dat['Species'], dat['size'])

# 9. Mosaic-like Plot (stacked bar instead of mosaic)
cont_table.plot(kind="bar", stacked=True, colormap="Set2")
plt.title("Species vs Size")
plt.ylabel("Count")
plt.show()

# 10. Additional Visualizations
plt.hist(dat['sepal length (cm)'], color="skyblue", bins=10, edgecolor="black")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

sns.boxplot(x="Species", y="sepal length (cm)", data=dat, palette="light:#5A9")
plt.title("Boxplot of Sepal Length by Species")
plt.show()

# Print all computed results
print("Minimum:", min_val)
print("Maximum:", max_val)
print("Range:", range_val, "\n")

print("Mean:", mean_val)
print("Median:", median_val, "\n")

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", iqr_val, "\n")

print("Variance:", var_val)
print("Standard Deviation:", sd_val, "\n")

print("Mode:", mode_val, "appears", mode_freq, "times\n")

print("Correlation between Sepal Length and Sepal Width:", cor_val, "\n")

print("Contingency Table:\n", cont_table, "\n")

print("Summary Statistics:\n", summary_stats)
```
