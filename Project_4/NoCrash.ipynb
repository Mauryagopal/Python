```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python


df=pd.read_csv('6accident_data.csv')
```

```python
#check columns
print("\n",df.columns)
print("\n",df.shape)
```

```python


sns.countplot(data=df, x='Accident_Severity')
plt.title("Accident Severity Distribution")
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
sns.countplot(data=df, x='Year')
plt.title("Accidents per Year")
plt.xlabel("Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(data=df, y='Weather_Conditions', order=df['Weather_Conditions'].value_counts().index)
plt.title("Accidents by Weather Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condition")
plt.show()

```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Road_Type', y='Speed_limit')
plt.title("Speed Limit Distribution across Road Types")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x='Number_of_Vehicles', y='Number_of_Casualties', hue='Accident_Severity')
plt.title("Casualties vs. Vehicles in Accidents")
plt.xlabel("Number of Vehicles")
plt.ylabel("Number of Casualties")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap

m = folium.Map(location=[df["Latitude"].mean(), df["Longitude"].mean()], zoom_start=6)

heat_data = [[row['Latitude'], row['Longitude']] for index, row in df.iterrows()]
HeatMap(heat_data, radius=8).add_to(m)

m.save("accident_heatmap.html")
```
