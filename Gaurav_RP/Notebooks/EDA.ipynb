## Color parameter evolution

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,4))
for i,c in enumerate(["L*","a*","b*"]):
    plt.subplot(1,3,i+1)
    sns.boxplot(x="Date",y=c,data=df,palette="viridis")
    plt.title(f"{c} over time")
plt.tight_layout(); plt.show()
```

## 1️⃣ L* (Lightness)

The median L* value decreases from Day 1 → Day 3, showing that the cake becomes progressively darker.
The spread also narrows slightly, meaning color darkening is happening consistently across samples.
A few lower outliers suggest localized over‑browning or uneven surface darkening.

2️⃣ a* (Red–Green axis)

a* values shift slightly downward (more negative) with time.
This means the red tone is fading and the product is moving slightly toward a greenish tint — a subtle hue change often observed as surfaces lose their reddish freshness.
Variation remains moderate, so the change is not the dominant one.

b* (Yellow–Blue axis)

b* values show a steady and strong increase through Day 3.
This is the most pronounced trend: surfaces develop a higher yellow or brown component, characteristic of Maillard browning or oxidative color formation.
Higher outliers on Day 3 confirm some samples browned more rapidly than others.

```python
rgb_cols = ["R", "G", "B"]
df.groupby("Date")[rgb_cols].mean().plot(kind="bar", figsize=(8,4))
plt.title("Average RGB values over time")
plt.ylabel("Mean intensity")
plt.show()
```

1️⃣ Red (R)

The red channel intensity stays high across all days, with only a small decline by Day 3.
This suggests the surface retains its reddish component for most of the period, then slightly loses brightness due to darkening.

2️⃣ Green (G)

The green channel follows a similar pattern to red, staying fairly stable with a minor decrease on Day 3.
Together, red + green dominance means the cake’s surface tone remains in the yellowish region, consistent with baked color tones.

Blue (B)

Blue intensity drops significantly by Day 3.
Reduction in blue reflectance makes the color visually warmer (more yellow/brown), a common result of browning reactions or surface oxidation.

```python
wavelength_cols = [col for col in df.columns if "nm" in col]
print(f"Spectral range: {wavelength_cols[0]}–{wavelength_cols[-1]}")
```

```python
plt.figure(figsize=(10,6))
for date, sub in df.groupby("Date"):
    plt.plot(range(len(wavelength_cols)),
             sub[wavelength_cols].mean(),
             label=date)
plt.xlabel("Wavelength index (approx 340–850 nm)")
plt.ylabel("Mean reflectance/absorbance")
plt.title("Spectral curve over time")
plt.legend()
plt.show()
```

## Spectral Behaviour Insights
1️⃣ Broad spectral peak pattern

The curve rises from near‑zero reflectance around 340 nm, reaches a strong peak between 450 – 650 nm, and then gradually declines again toward 850 nm.
This shape matches typical reflectance behavior of baked surfaces, which reflect most visible light except in specific absorbance regions linked to natural pigments and browning products.

2️⃣ Slight downward shift with time (Day 1 → Day 3)

The Day 3 curve lies marginally below Day 1, especially in the 400 – 500 nm region (blue wavelength range).
This indicates increasing absorption of blue light — the key spectral signature of yellow/brown color formation.
Even small dips here are significant because human perception is very sensitive in the blue region.

3️⃣ Stable reflectance in mid – red region (550–700 nm)

Reflectance values remain similar across days in this region, suggesting no major gain in red coloration, only darkening through diminished blue reflectance.


4️⃣ Near‑infrared region (700 – 850 nm)

Reflectance remains steady and relatively low, implying that moisture content or texture changes (which show most strongly in NIR) are minimal.
The dominant change is therefore surface pigment chemistry, not structural or water‑loss related.


```python
trend = df.groupby("Date")[["L*", "a*", "b*"]].mean()
trend.plot(marker='o', figsize=(7,4))
plt.title("Average color evolution over storage period")
plt.ylabel("Mean value")
plt.show()
```

## Color Trend Insights (L*, a*, b*)
1️⃣ Lightness (L*)

Mean L* values rise slightly from Day 1 → Day 2, then decrease again by Day 3.
This small drop at the end signals darkening over time, consistent with ongoing browning reactions as the storage period extends.
The Day 2 plateau suggests the change began subtly before becoming visible later.

2️⃣ Red–Green axis (a*)

a* values remain slightly negative and nearly constant throughout.
This shows minimal change along the red‑green spectrum — the degradation process primarily impacts brightness and yellowness, not redness.

3️⃣ Yellow–Blue axis (b*)

b* increases clearly from Day 1 → Day 3.
The rise indicates an intensifying yellow–brown hue, the classic signature of Maillard or oxidative browning in baked goods.
The magnitude of this increase is larger than other coordinates, confirming it as the dominant color shift.

The combined colorimetric (L*, a*, b*) and spectral analyses reveal consistent, physically meaningful patterns of surface darkening and yellow‑browning, characteristic of food product degradation during storage.
