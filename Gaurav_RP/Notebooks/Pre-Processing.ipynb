##  Outlier detection (color features)

```python
import seaborn as sns
import matplotlib.pyplot as plt
color_cols = ["L*", "a*", "b*"]

plt.figure(figsize=(10,3))
for i,c in enumerate(color_cols):
    plt.subplot(1,3,i+1)
    sns.boxplot(data=df, x="Date", y=c, palette="Set2")
    plt.title(c)
plt.tight_layout()
plt.show()
```

## The collective pattern L* ↓ + a* ≈ constant + b* ↑ confirms progressive yellow‑browning and darkening — typical of baked or stored foods undergoing pigment formation or oxidation during storage.

L** has a few low‑value outliers (very dark samples).

a** shows a couple of points much lower or higher than the main group.

b** has a handful that are either unusually low or high relative to the batch.

Those mild outliers are not necessarily bad — they can be scientifically meaningful:

Very low L* and low b* points → dark patches, overbaked regions.

Moderate high b* points → early caramelization or brighter crumb areas.

##  IQR bounds for L*, a*, b* and create outlier flags

```python
import numpy as np
import pandas as pd

def flag_outliers(series, k=1.5):
    """Return a boolean Series marking outliers"""
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    low, high = q1 - k*iqr, q3 + k*iqr
    return (series < low) | (series > high)

for col in ["L*", "a*", "b*"]:
    df[f"{col}_outlier"] = flag_outliers(df[col])
```

```python
outlier_summary = df[["L*_outlier","a*_outlier","b*_outlier"]].sum().to_frame("count")
outlier_summary["percent"] = 100 * outlier_summary["count"] / len(df)
print(outlier_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,3))
for i, c in enumerate(["L*","a*","b*"]):
    plt.subplot(1,3,i+1)
    sns.scatterplot(x="Date", y=c, data=df,
                    hue=f"{c}_outlier", palette={True:"r", False:"gray"},
                    alpha=0.7)
    plt.title(f"{c} with Outlier Flag")
plt.tight_layout()
plt.show()
```
