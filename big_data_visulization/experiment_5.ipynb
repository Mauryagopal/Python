```python
import pandas as pd
import matplotlib.pyplot as plt 

# Define the data
series1 = pd.Series([12, 24, 36, 48, 60, 72, 84])
series2 = pd.Series([2, 4, 3, 18, 60, 72, 84])

# Plot the data
plt.plot(
    series1,
    series2,
    color='green',          # Line color
    linestyle='dashed',     # Line style
    linewidth=3,            # Line thickness
    marker='o',             # Marker type
    markerfacecolor='blue', # Marker fill color
    markersize=12           # Marker size
)

# Add labels and title
plt.title("Line Plot of Series1 vs Series2")
plt.xlabel("Series1 Values")
plt.ylabel("Series2 Values")

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Number of rows
n = 10

# Generate random data
time_period = np.arange(2000, 2000 + n*5, 5)  # e.g., 2000, 2005, 2010,...
population = np.random.randint(1000000000, 1400000000, size=n)  # Random population
gdp = np.random.randint(400, 3000, size=n)  # Random GDP in billions

# Create DataFrame
data = pd.DataFrame({
    "Time period": time_period,
    "Population": population,
    "GDP": gdp
})

# Save to CSV
data.to_csv("CountryData.IND.csv", index=False)

# Show the first few rows
print(data)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Number of data points
n = 1000

# Generate normally distributed data
# Mean = 50, Standard Deviation = 10
time_period = np.random.normal(loc=50, scale=10, size=n)

# Optional: create a DataFrame
data = pd.DataFrame({
    "Time period": time_period
})

# Plot histogram
plt.hist(data['Time period'], bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normally Distributed Time Period")
plt.xlabel("Time period")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate date range (like your example 2014-06 to 2018-08)
dates = pd.date_range(start='2014-06-01', end='2018-08-31', freq='B')  # 'B' = business days

# Initialize stock price
price = 200  # starting price
prices = []

# Generate stock prices using random walk
for _ in range(len(dates)):
    daily_return = np.random.normal(loc=0.0005, scale=0.02)  # small drift and daily volatility
    price *= (1 + daily_return)
    prices.append(price)

# Create DataFrame
df = pd.DataFrame({'Close': prices}, index=dates)

# Plot the simulated stock prices
df.plot(title="Simulated Stock Prices (AMZN-like)")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```
